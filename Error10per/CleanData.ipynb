# **Data Cleaning**

```python
new_colors = {"D": 0, "E": 1,"F": 2,"G": 3,"H": 4,"I": 5,"J": 6}        # Create colors dictoinary for map key and value
data["color"].replace(new_colors, inplace=True)                         # Map 
data.head()
```

```python
new_clarity = {"FL":0,  "IF": 1, "VVS1": 2, "VVS2": 3, "VS1": 4, "VS2": 5, "SI1": 6, "SI2": 7, "SI3": 8, "I1": 9}
data["clarity"].replace(new_clarity, inplace=True)
data.head()
```

```python
new_cut = {"Ideal": 0, "Premium": 1, "Very Good": 2, "Good": 3, "Fair": 4}
data["cut"] .replace(new_cut, inplace=True)
data.head()
```

# **Drop Row**

# Drop X

```python
dropList = list(data[(data["x"] < 2)  | (data["x"] > 6)].index)
data = data.drop(dropList)

outlier = list(data[((data["x"] < 5) & (data["price"] > 1500))].index)
data = data.drop(outlier)

outlier = list(data[((data["x"] > 5.4) & (data["price"] < 1500))].index)
data = data.drop(outlier)

outlier = list(data[((data["x"] < 4.1) & (data["price"] > 900))].index)
data = data.drop(outlier)
```

# Drop Y

```python
dropList = list(data[data["y"] >= 5.8].index)
data = data.drop(dropList)

outlier = list(data[((data["y"] < 2.4) & (data["price"] > 1500))].index)
data = data.drop(outlier)
```

# Drop Z

```python
dropList = list(data[(data["z"] < 2.25) | (data["z"] > 3.6)].index)
data = data.drop(dropList)

outlier = list(data[((data["z"] < 2.4) & (data["price"] > 1500))].index)
data = data.drop(outlier)

outlier = list(data[((data["z"] < 0.3) & (data["price"] > 1000))].index)
data = data.drop(outlier)

outlier = list(data[((data["z"] < 0.3) & (data["price"] > 1000))].index)
data = data.drop(outlier)
```

# Drop Carat

```python
dropList = list(data[data["carat"] >= 0.75].index)
data = data.drop(dropList)

outlier = list(data[((data["carat"] < 0.3) & (data["price"] > 1000))].index)
data = data.drop(outlier)

outlier = list(data[((data["carat"] > 0.5) & (data["price"] < 1000))].index)
data = data.drop(outlier)
```

# Drop Table

```python
dropList = list(data[(data["table"] < 50) | (data["table"] > 66)].index)
data = data.drop(dropList)
```

# Drop Depth

```python
dropList = list(data[(data["depth"] < 59) | (data["depth"] > 64)].index)
data = data.drop(dropList)
```

# Drop Price

```python
sd=data["price"].std()
price_mean=data["price"].mean()
price_med=data["price"].median()
print("S.D. =",sd)
print("Mean =",price_mean)
print("Med =",price_med)

z3=(1*sd)+price_mean
print("Z3 =",z3)
```

```python
dropList = list(data[data["price"] > z3].index)
data = data.drop(dropList)
```
